```python
import pandas as pd

df = pd.read_csv("SALES.csv")
df.head()
```

```python
# Basic info
df.info()

# Check for null values
df.isnull().sum()

# Summary statistics
df.describe()
```

```python
total_sales = df['Sales'].sum()
print("Total Sales:",total_sales)
```

```python
sales_by_product = df.groupby('Product')['Sales'].sum()
print(sales_by_product)
```

```python
sales_by_month = df.groupby('Month')['Sales'].sum()
print(sales_by_month)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values, palette="viridis")
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=sales_by_month.index, y=sales_by_month.values, palette="Set2")
plt.title("Total Sales by Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='Month', y='Sales', hue='Product', marker="o")
plt.title("Monthly Sales Trend per Product")
plt.ylabel("Sales")
plt.show()
```

```python
# Top product overall
top_product = sales_by_product.idxmax()
print("Top Product Overall:", top_product)

# Month with highest sales
top_month = sales_by_month.idxmax()
print("Month with Highest Sales:",top_month)
```
